# Gráficos

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math as math
```

```python
def correlacion_pearson(data_frame, columna1, columna2):
    return np.corrcoef(data_frame[columna1], data_frame[columna2])[0,1];
```

### Experimento 1: Fuerza Bruta
La hipótesis es que fuerza bruta se comporta de la misma forma para todas las familias de instancias.

```python
data_pc = pd.read_csv("experimentos/resultado_pc_fb.csv")
data_mc = pd.read_csv("experimentos/resultado_mc_fb.csv")
data_pc["tiempo_esperado"] = (data_pc["n"] * (2 ** data_pc["n"]) ) * 0.0000003;
data_pc["complejidad"] = data_pc["n"] * (2 ** data_pc["n"]) ;
```

```python
# Comparación peor-mejor caso
fig = sns.scatterplot(data=data_pc, x="n", y="tiempo", color="blue");
fig = sns.lineplot(data=data_mc, x='n', y='tiempo', color="red");

fig.legend(labels=['FB - Peor Caso','FB - Mejor Caso']);
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/fb-comparacion-peor-mejor-caso.svg");  
```

```python
# Comparación con n * 2^n
fig = sns.scatterplot(data=data_pc, x="n", y="tiempo", color="green");
fig = sns.lineplot(data=data_pc, x='n', y='tiempo_esperado', color="black");

fig.legend(labels=['n * 2^n','FB']);
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/fb-complejidad.svg");
```

```python
# Correlación de Pearson
fig = sns.lmplot(data=data_pc, x="tiempo", y="complejidad");
plt.savefig("graficos/bt-correlacion-peor-caso.svg");
print("Índice de correlación de Pearson:", correlacion_pearson(data_pc, "tiempo", "complejidad"));
```

### Experimento 2: Backtracking (peor caso)
La hipotesis es que backtracking va a tener comportamiento exponencial en peor caso. Y lineal en el mejor caso

```python
data = pd.read_csv("experimentos/resultado_pc_bt.csv")
data["tiempo_esperado"] = (data["n"] * (2 ** data["n"]) ) * 0.0000003;
data["complejidad"] = data["n"] * (2 ** data["n"]);

# Comparación con fn teórica
fig = sns.scatterplot(data=data, x="n", y="tiempo");
fig = sns.lineplot(data=data, x='n', y='tiempo_esperado', color="black");
fig.legend(labels=['O(n * 2^n)','BT']);
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/bt-complejidad-peor-caso.svg");  

# Correlación de Pearson
fig = sns.lmplot(data=data, x="tiempo", y="complejidad");
plt.savefig("graficos/bt-correlacion-peor-caso.svg");
print("Índice de correlación de Pearson:", correlacion_pearson(data, "tiempo", "complejidad"));
```

### Experimento 4: Programación Dinamica

```python
Ms = [4000, 6000, 8000, 10000]
colors = ["#70a3d0", "#b75151", "#e8bc3d", "#51b788"]

fig, axs = plt.subplots(2, 2);
fig.set_figwidth(13);
fig.set_figheight(10);

# Gráfico 1
color = 0;
labels = [];
for M in Ms:
    data = pd.read_csv(F"experimentos/resultado_pd_DCN-{M}.csv");
    fig = sns.lineplot(data=data, x="n", y="tiempo", color=colors[color], alpha=1, ax=axs[0][0]);
    labels.append(F"M={M}");
    color += 1;

data["tiempo_esperado"] = data["n"] * 0.07;
fig = sns.lineplot(data=data, x="n", y="tiempo_esperado", color="black", alpha=1, ax=axs[0][0]);
labels.append(F"O(n)");
    
axs[0][0].set_title("Tiempo por cantidad de locales");
axs[0][0].set(xlabel='n', ylabel='tiempo (ms)');
axs[0][0].legend(labels=labels);

# Gráfico 2
color = 0
for M in Ms:
    data = pd.read_csv(F"experimentos/resultado_pd_DCN-{M}.csv")
    fig = sns.lineplot(data=data, x="n", y="tiempo", color=colors[color], alpha=0.5, ax=axs[0][1]);
    labels.append(F"M={M}")
    color += 1
    
axs[0][1].set_title("Tiempo por cantidad de locales (alpha=0.5)");
axs[0][1].set(xlabel='n', ylabel='tiempo (ms)');
    
# Gráfico 3    
color = 0;
labels = [];
for M in Ms:
    data = pd.read_csv(F"experimentos/resultado_pd_DC-{M}.csv")
    fig = sns.lineplot(data=data, x="n", y="tiempo", color=colors[color], alpha=1, ax=axs[1][0]);
    labels.append(F"n={M}");
    color += 1

data["tiempo_esperado"] = data["n"] * 0.07;
fig = sns.lineplot(data=data, x="n", y="tiempo_esperado", color="black", alpha=1, ax=axs[1][0]);
labels.append(F"O(M)");    

axs[1][0].set_title("Tiempo por límite de contagio");
axs[1][0].set(xlabel='M', ylabel='tiempo (ms)');
axs[1][0].legend(labels=labels);

# Gráfico 4
color = 0
for M in Ms:
    data = pd.read_csv(F"experimentos/resultado_pd_DC-{M}.csv")
    fig = sns.lineplot(data=data, x="n", y="tiempo", color=colors[color], alpha=0.5, ax=axs[1][1]);
    color += 1
    
axs[1][1].set_title("Tiempo por límite de contagio (alpha=0.5)");
axs[1][1].set(xlabel='M', ylabel='tiempo (ms)');


plt.savefig("graficos/pd.svg");
```

#### Experimento 5: Bandas de contagio

```python
# Comparación con fn teórica
ms = [10, 100, 1000]
Ms = [10000]

fig, axs = plt.subplots(1, 2);
fig.set_figwidth(16);

for i in range(0, len(Ms)):
    axs[i].set_title(F"M={Ms[i]}");
    data = []
    labels = []
    colors = ["#70a3d0", "#b75151", "#e8bc3d", "#51b788", "#6750c3"]
    color = 0
    for m in ms:
        data = pd.read_csv(F"experimentos/resultado_pd_BC-{Ms[i]}-{m}.csv")
        fig = sns.lineplot(data=data, x="n", y="tiempo", color=colors[color], ax=axs[i]);
        labels.append(F"Rango=[1,{m}]")
        color += 1

fig.legend(labels=labels);
fig.set(xlabel='n', ylabel='tiempo (ms)');
#fig.set_title(F"M={M}");
plt.savefig(F"graficos/pd_bc.svg");
```

#### Experimento 7: Comparación podas BT con 4 casos generales

```python
PODAS = ['BT', 'BT-F', 'BT-O']
colors = ["#70a3d0", "#b75151", "#e8bc3d", "#51b788"]
bMaxs = ["CONSTANTE", "INCREMENTAL"]
nivelesContagio = ["BAJO", "ALTO"]

fig, axs = plt.subplots(2, 2);
fig.set_figwidth(13);
fig.set_figheight(10);
# plt.yscale("log");

# Gráficos
color = 0;
labels = [];
for i in range(0, len(bMaxs)):
    for j in range(0, len(nivelesContagio)):
        color = 0;
        for poda in PODAS:
            data = pd.read_csv(F"experimentos/resultado_{poda}-PODA-{bMaxs[i]}-{nivelesContagio[j]}.csv");
            fig = sns.lineplot(data=data, x="n", y="tiempo", color=colors[color], alpha=1, ax=axs[i][j]);
            labels.append(F"PODA={poda}");
            color += 1;
            axs[i][j].set_title(F"Nivel de contagio {nivelesContagio[j]} y beneficio {bMaxs[i]}");
            axs[i][j].set(xlabel='n', ylabel='tiempo (ms)');
            axs[i][j].legend(labels=labels);
#             axs[i][j].set_yscale('log');

# data["tiempo_esperado"] = data["n"] * 0.07;
# fig = sns.lineplot(data=data, x="n", y="tiempo_esperado", color="black", alpha=1, ax=axs[0][0]);
# labels.append(F"O(n)");

plt.savefig("graficos/EXP-7.svg");
```
